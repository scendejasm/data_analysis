```python
import pandas as pd
import numpy as np
```

## Importing data from Mainframe query of all related GUID

```python
capi_cor_df = pd.read_csv('CAPI1_correspondence_GUID.csv')
```

```python
capi_cor_df.columns = capi_cor_df.columns.str.lower()
```

```python
capi_cor_df.head()
```

## Importing data from Mainframe query of all related Parcels 

```python
capp_cor_df = pd.read_csv('CAPP1_correspondence_GUID.csv')
```

```python
capp_cor_df.head()
```

### Change column headers to lowercase and clean up the parcel numbers  

```python
capp_cor_df.columns = capp_cor_df.columns.str.lower()
```

```python
capp_cor_df['parcel_nbr_int'] = capp_cor_df['capp_parcel_nbr'].astype(int)
```

```python
capp_cor_df.head()
```

```python
capp_cor_df['parcel_nbr_clean'] = capp_cor_df['parcel_nbr_int'].astype(str).apply(lambda x: '0' + x if len(x) == 12 else x)
```

```python
capp_cor_df.head()
```

## Merge the CAPI DataFram with CAPP based on matching GUID

```python
merged_cap_cor_df = pd.merge(capp_cor_df, capi_cor_df, left_on='capp_guid', right_on='capi_guid', how='outer')
```

```python
merged_cap_cor_df.head()
```

## Get only related columns and covert date to a datetime

```python
cap_cor_df = merged_cap_cor_df[['capp_guid', 'parcel_nbr_clean', 'capi_event_date']]
```

```python
cap_cor_df.loc[:,'event_date'] = pd.to_datetime(cap_cor_df['capi_event_date'])
cap_cor_df.head()
```
