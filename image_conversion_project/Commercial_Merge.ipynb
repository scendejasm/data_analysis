# Merge of Commercial Appraisals extracts

```python
import pandas as pd
import numpy as np
```

## Import of Parcel info from Mainframe CAPP1

```python
capp_com_df = pd.read_csv('CAPP1_commercial_GUID.csv')
```

```python
capp_com_df.head()
```

### Clean headers, and parcel number

```python
capp_com_df.columns = capp_com_df.columns.str.lower()
```

```python
capp_com_df['parcel_nbr_int'] = capp_com_df['capp_parcel_nbr'].astype(int)
```

```python
capp_com_df.head()
```

```python
capp_com_df['parcel_nbr_clean'] = capp_com_df['parcel_nbr_int'].astype(str).apply(lambda x: '0' + x if len(x) == 12 else x)
```

```python
capp_com_df.head()
```

## Import data from Mainframe query of all related GUID

```python
capi_com_df = pd.read_csv('CAPI1_commercial_GUID.csv')
```

```python
capi_com_df.columns = capi_com_df.columns.str.lower()
```

```python
capi_com_df.head()
```

```python
capi_com_df.loc[:,'event_date'] = pd.to_datetime(capi_com_df['capi_event_date'])
```

```python
capi_com_df.info()
```

## Merge CAPP df with CAPI on GUID

```python
merged_com_df = pd.merge(capp_com_df, capi_com_df, left_on='capp_guid', right_on='capi_guid', how='outer')
```

```python
merged_com_df.head()
```

## Get only relevant columns and do any additional cleaning

```python
mainframe_df = merged_com_df[['capp_guid','parcel_nbr_clean','event_date']]
```

```python
mainframe_df.info()
```

## Import the Images to be Deleted list from Dawn Solis

```python
dawn_com_df = pd.read_csv('image_list_commercial_dawn.csv')
```

```python
dawn_com_df.head()
```

### Clean Parcel number and convert date to datetime

```python
dawn_com_df['parcel_nbr_clean'] = dawn_com_df['parcel_nbr'].str.replace('-','')
```

```python
dawn_com_df['event_date'] = pd.to_datetime(dawn_com_df['event_date'])
```

```python
dawn_com_df['event_date2'] = pd.to_datetime(dawn_com_df['event_date2'])
```

```python
dawn_com_df.info()
```

## Finally: Merge the Mainframe table with Dawn's table on Parcel # and Event Date 

```python
dawn_com_merged_df = pd.merge(dawn_com_df, mainframe_df, on=['parcel_nbr_clean', 'event_date'], how='inner')
```

```python
dawn_com_merged_df
```

```python
dawn_com_merged_df2 = pd.merge(dawn_com_df, mainframe_df, left_on=['parcel_nbr_clean', 'event_date2'], right_on=['parcel_nbr_clean', 'event_date'],how='inner')
```

```python
dawn_com_merged_df2
```

```python
final_merged_df = pd.concat([dawn_com_merged_df, dawn_com_merged_df2])
```

```python
final_merged_df
```

## Output the results to share 

```python
final_merged_df.to_csv('steves_findings_guid_list.csv', index=False) 
```
